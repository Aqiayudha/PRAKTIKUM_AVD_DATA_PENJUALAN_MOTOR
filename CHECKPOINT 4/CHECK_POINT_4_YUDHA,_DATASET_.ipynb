### **Outliers Values**

Pengecekan outliers dilakukan untuk melihat apakah ada data yang nilainya menyimpang jauh dari rata-rata kategori lainnya pada kolom numerik (HARGA JUAL dan DISKON).

```python
import pandas as pd

# Memuat ulang DataFrame dengan pemisah ';' untuk mendapatkan kolom-kolom terpisah.
# DataFrame df yang ada saat ini hanya memiliki satu kolom 'object'.
file = '/content/drive/MyDrive/DataPenjualanMotor (2).csv'
df = pd.read_csv(file, sep=';')

# Membersihkan kolom 'HARGA JUAL' dan 'DISKON' dari karakter non-numerik sebelum konversi.
# Asumsi: Titik '.' dan koma ',' mungkin digunakan sebagai pemisah ribuan dan harus dihapus.
# Pertama, pastikan kolom bertipe string agar bisa menggunakan metode .str.replace()
df['HARGA JUAL'] = df['HARGA JUAL'].astype(str).str.replace('.', '', regex=False).str.replace(',', '', regex=False)
df['DISKON'] = df['DISKON'].astype(str).str.replace('.', '', regex=False).str.replace(',', '', regex=False)

# Mengonversi kolom 'HARGA JUAL' dan 'DISKON' ke tipe data numerik.
# 'errors='coerce'' akan mengubah nilai yang tidak bisa dikonversi menjadi NaN (Not a Number).
df['HARGA JUAL'] = pd.to_numeric(df['HARGA JUAL'], errors='coerce')
df['DISKON'] = pd.to_numeric(df['DISKON'], errors='coerce')

# Menghapus baris yang memiliki nilai NaN di kolom 'HARGA JUAL' atau 'DISKON'
# setelah konversi, karena NaN akan mengganggu perhitungan outlier.
df.dropna(subset=['HARGA JUAL', 'DISKON'], inplace=True)

results = []

# Sekarang, `df.select_dtypes` akan dapat menemukan kolom numerik yang benar.
cols_to_analyze = df.select_dtypes(include=['float64', 'int64']).columns

if cols_to_analyze.empty:
    print("Peringatan: Tidak ada kolom numerik yang ditemukan untuk analisis outlier setelah persiapan data.")
elif len(df) == 0:
    # Menambahkan kondisi ini untuk menangani kasus di mana df menjadi kosong setelah dropna
    print("Peringatan: DataFrame kosong setelah membersihkan nilai non-numerik. Tidak dapat menghitung outlier.")
else:
    for col in cols_to_analyze:
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        percent_outliers = (len(outliers) / len(df)) * 100
        results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Membuat DataFrame dari list hasil
results_df = pd.DataFrame(results)

# Hanya mengatur indeks jika results_df tidak kosong
if not results_df.empty:
    results_df.set_index('Kolom', inplace=True)
    # Menghapus nama indeks agar tampilannya lebih rapi
    results_df.index.name = None
else:
    # Jika tidak ada kolom numerik yang ditemukan atau hasil kosong,
    # buat DataFrame kosong dengan header kolom yang diharapkan.
    results_df = pd.DataFrame(columns=['Persentase Outliers'])

# Menampilkan DataFrame
display(results_df)
```

Berdasarkan metode IQR (Interquartile Range), sekitar 1,32% data pada kolom HARGA JUAL teridentifikasi sebagai outlier.

```python
import pandas as pd
file = '/content/drive/MyDrive/DataPenjualanMotor (2).csv'
df = pd.read_csv(file, sep=';')

df['HARGA JUAL'] = df['HARGA JUAL'].astype(str).str.replace('.', '', regex=False).str.replace(',', '', regex=False)
df['DISKON'] = df['DISKON'].astype(str).str.replace('.', '', regex=False).str.replace(',', '', regex=False)

df['HARGA JUAL'] = pd.to_numeric(df['HARGA JUAL'], errors='coerce')
df['DISKON'] = pd.to_numeric(df['DISKON'], errors='coerce')
df.dropna(subset=['HARGA JUAL', 'DISKON'], inplace=True)

def kategorikan_harga_jual(harga_value):
    if harga_value > 30000000:
        return 'Harga Tinggi'
    elif harga_value < 20000000:
        return 'Harga Rendah'
    else:
        return 'Harga Sedang'

df['KETERANGAN HARGA'] = df['HARGA JUAL'].apply(kategorikan_harga_jual)
```
