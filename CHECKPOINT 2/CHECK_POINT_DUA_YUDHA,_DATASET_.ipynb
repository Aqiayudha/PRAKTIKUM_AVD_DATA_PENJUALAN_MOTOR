### **Outliers Values**

Pengecekan outliers dilakukan untuk melihat apakah ada data yang nilainya menyimpang jauh dari rata-rata kategori lainnya pada kolom numerik (HARGA JUAL dan DISKON).

```python
import pandas as pd

# Memuat ulang DataFrame dengan pemisah ';' untuk mendapatkan kolom-kolom terpisah.
# DataFrame df yang ada saat ini hanya memiliki satu kolom 'object'.
file = '/content/drive/MyDrive/DataPenjualanMotor (2).csv'
df = pd.read_csv(file, sep=';')

# Membersihkan kolom 'HARGA JUAL' dan 'DISKON' dari karakter non-numerik sebelum konversi.
# Asumsi: Titik '.' dan koma ',' mungkin digunakan sebagai pemisah ribuan dan harus dihapus.
# Pertama, pastikan kolom bertipe string agar bisa menggunakan metode .str.replace()
df['HARGA JUAL'] = df['HARGA JUAL'].astype(str).str.replace('.', '', regex=False).str.replace(',', '', regex=False)
df['DISKON'] = df['DISKON'].astype(str).str.replace('.', '', regex=False).str.replace(',', '', regex=False)

# Mengonversi kolom 'HARGA JUAL' dan 'DISKON' ke tipe data numerik.
# 'errors='coerce'' akan mengubah nilai yang tidak bisa dikonversi menjadi NaN (Not a Number).
df['HARGA JUAL'] = pd.to_numeric(df['HARGA JUAL'], errors='coerce')
df['DISKON'] = pd.to_numeric(df['DISKON'], errors='coerce')

# Menghapus baris yang memiliki nilai NaN di kolom 'HARGA JUAL' atau 'DISKON'
# setelah konversi, karena NaN akan mengganggu perhitungan outlier.
df.dropna(subset=['HARGA JUAL', 'DISKON'], inplace=True)

results = []

# Sekarang, `df.select_dtypes` akan dapat menemukan kolom numerik yang benar.
cols_to_analyze = df.select_dtypes(include=['float64', 'int64']).columns

if cols_to_analyze.empty:
    print("Peringatan: Tidak ada kolom numerik yang ditemukan untuk analisis outlier setelah persiapan data.")
elif len(df) == 0:
    # Menambahkan kondisi ini untuk menangani kasus di mana df menjadi kosong setelah dropna
    print("Peringatan: DataFrame kosong setelah membersihkan nilai non-numerik. Tidak dapat menghitung outlier.")
else:
    for col in cols_to_analyze:
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        percent_outliers = (len(outliers) / len(df)) * 100
        results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Membuat DataFrame dari list hasil
results_df = pd.DataFrame(results)

# Hanya mengatur indeks jika results_df tidak kosong
if not results_df.empty:
    results_df.set_index('Kolom', inplace=True)
    # Menghapus nama indeks agar tampilannya lebih rapi
    results_df.index.name = None
else:
    # Jika tidak ada kolom numerik yang ditemukan atau hasil kosong,
    # buat DataFrame kosong dengan header kolom yang diharapkan.
    results_df = pd.DataFrame(columns=['Persentase Outliers'])

# Menampilkan DataFrame
display(results_df)
```

# **Eksplorasi Data (EDA)**

## **Comparison/Perbandingan**

Analisis perbandingan dilakukan untuk melihat hubungan antar variabel dan pola transaksi pada dataset Penjualan Motor.

- Aktivitas: Melakukan komparasi total transaksi yang berasal dari berbagai kanal penjualan yang tercatat pada kolom SUMBER dalam dataset.

- Tujuan: Menentukan efektivitas saluran penjualan untuk melihat apakah pelanggan lebih banyak melakukan pembelian secara langsung (Offline/Showroom) atau melalui media digital (Online/Social Media).

- Visualisasi: Menggunakan Bar Chart (Grafik Batang).

```python
penjualan_market = df.groupby('SUMBER')['HARGA JUAL'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=penjualan_market.index, y=penjualan_market.values, palette='viridis', hue=penjualan_market.index, legend=False)
plt.title('Total Penjualan Berdasarkan Sumber (Online/Offline)')
plt.xlabel('Sumber Penjualan')
plt.ylabel('Total Penjualan (HARGA JUAL)')
plt.xticks(rotation=45)
plt.show()
```

Berdasarkan analisis pada diagram batang yang telah dibuat, berikut adalah beberapa poin penting yang dapat disimpulkan:

- Dominasi Saluran Penjualan: Saluran penjualan Showroom (Offline) masih menjadi penyumbang terbesar dalam total transaksi dibandingkan dengan saluran online.

- Kepercayaan Pelanggan: Tingginya angka pada kategori offline menunjukkan bahwa dalam pembelian unit kendaraan bermotor, pelanggan masih memiliki preferensi kuat untuk melihat unit secara fisik dan bertransaksi langsung di lokasi.

- Potensi Digital: Meskipun angka offline lebih dominan, terdapat kontribusi yang signifikan dari sumber digital (seperti Social Media atau Website), yang menandakan bahwa proses pencarian informasi oleh konsumen sudah mulai beralih ke ranah online sebelum akhirnya melakukan pembelian.

## **Composition/Komposisi**

- Aktivitas: Melakukan analisis proporsi penjualan dengan mengambil tiga kategori produk pada kolom SERIES yang memiliki total penjualan tertinggi dibandingkan kategori lainnya dalam dataset.

- Tujuan: Mengetahui dominasi pasar dari tiga produk unggulan serta melihat seberapa besar persentase kontribusi "Top 3" tersebut terhadap keseluruhan total transaksi perusahaan.

- Visualisasi: Menggunakan Pie Chart (Diagram Lingkaran).

Analisis menggunakan diagram lingkaran ini memberikan gambaran mengenai seberapa besar dominasi seri produk tertentu terhadap total volume penjualan:

- Dominasi Produk Utama: Tiga seri produk teratas (seperti Matic High, Matic Low, dan Sport) menguasai sebagian besar pangsa pasar penjualan motor dalam dataset ini.

- Konsentrasi Pasar: Besarnya persentase pada "Top 3" menunjukkan bahwa preferensi konsumen sangat terfokus pada model-model tertentu, sehingga perputaran stok pada seri ini jauh lebih cepat dibandingkan seri lainnya.

## **Distribution/Distribusi**

Aktivitas: Melakukan analisis distribusi frekuensi pada kolom HARGA JUAL untuk melihat bagaimana persebaran nilai transaksi dari yang terendah hingga yang tertinggi dalam dataset.

Tujuan: Mengidentifikasi rentang harga motor yang paling sering terjual (modus) dan memahami pola daya beli konsumen, apakah terpusat pada harga ekonomis atau menyebar ke harga premium.

## **Relationship/Hubungan**

Aktivitas: Melakukan perhitungan koefisien korelasi antara variabel HARGA JUAL dan DISKON untuk melihat derajat hubungan linear di antara keduanya dalam dataset.

Tujuan: Mengetahui apakah besaran harga jual suatu unit motor mempengaruhi besaran diskon yang diberikan, serta memahami pola pemberian insentif harga pada kategori produk yang berbeda.

Visualisasi: Menggunakan Heatmap.

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data=df[['HARGA JUAL', 'DISKON']].corr(),
            annot=True,
            cmap='viridis',
            fmt='.2f')
plt.title('Korelasi antara Harga Jual dan Diskon')
plt.show()
```

Analisis korelasi melalui heatmap memberikan gambaran mengenai keterkaitan antar variabel numerik utama dalam strategi penetapan harga:

- Derajat Hubungan: Nilai korelasi yang ditampilkan menunjukkan sejauh mana perubahan pada HARGA JUAL diikuti oleh perubahan pada DISKON.

- Pola Kebijakan Diskon: Jika ditemukan korelasi positif, hal ini menandakan bahwa produk dengan harga jual yang lebih tinggi cenderung mendapatkan nominal diskon yang lebih besar untuk menarik minat pembeli pada segmen premium.

- Independensi Variabel: Jika nilai korelasi mendekati nol, berarti besaran diskon diberikan secara merata atau berdasarkan faktor lain (seperti promo musiman) dan tidak bergantung pada mahal atau murahnya harga motor tersebut.
